```python
from astropy.cosmology import WMAP5, WMAP7, WMAP9, Planck13, Planck15
```

```python
import numpy as np
from pylab import *
%matplotlib inline
```

```python
z = np.arange(0,0.3,0.1)
```

```python
plot(z, WMAP5.comoving_distance(z), label='wmap5')
plot(z, WMAP7.comoving_distance(z), label='wmap7')
plot(z, WMAP9.comoving_distance(z), label='wmap9')
plot(z, Planck13.comoving_distance(z), label='Planck13')
plot(z, Planck15.comoving_distance(z), label='Planck15')
xlabel('z')
ylabel('comoving distance (Mpc)')
legend(loc=2)
```

```python
dd = Planck15.comoving_distance(z)
plot(z, dd-WMAP5.comoving_distance(z), label='wmap5')
plot(z, dd-WMAP7.comoving_distance(z), label='wmap7')
plot(z, dd-WMAP9.comoving_distance(z), label='wmap9')
plot(z, dd-Planck13.comoving_distance(z), label='Planck13')
plot(z, dd-Planck15.comoving_distance(z), label='Planck15')
xlabel('z')
ylabel('[experiment - Planck15]comoving distance (Mpc)')
legend(loc=2)
```

```python
dd = Planck15.angular_diameter_distance(z)
plot(z, dd-WMAP5.angular_diameter_distance(z), label='wmap5')
plot(z, dd-WMAP7.angular_diameter_distance(z), label='wmap7')
plot(z, dd-WMAP9.angular_diameter_distance(z), label='wmap9')
plot(z, dd-Planck13.angular_diameter_distance(z), label='Planck13')
plot(z, dd-Planck15.angular_diameter_distance(z), label='Planck15')
xlabel('z')
ylabel('[experiment - Planck15] angular diameter distance (Mpc)')
legend(loc=2)
```

```python
dd = Planck15.angular_diameter_distance(z)
plot(z, (dd-WMAP5.angular_diameter_distance(z))/dd, label='wmap5')
plot(z, (dd-WMAP7.angular_diameter_distance(z))/dd, label='wmap7')
plot(z, (dd-WMAP9.angular_diameter_distance(z))/dd, label='wmap9')
plot(z, (dd-Planck13.angular_diameter_distance(z))/dd, label='Planck13')
xlabel('z')
ylabel('(experiment - Planck15)/Planck15 angular diameter distance (Mpc)')
legend()
```

```python
plot(z, WMAP5.angular_diameter_distance(z), label='wmap5')
plot(z, WMAP7.angular_diameter_distance(z), label='wmap7')
plot(z, WMAP9.angular_diameter_distance(z), label='wmap9')
plot(z, Planck13.angular_diameter_distance(z), label='Planck13')
plot(z, Planck15.angular_diameter_distance(z), label='Planck15')
xlabel('z')
ylabel('angular diameter distance (Mpc)')
legend(loc=2)
```
